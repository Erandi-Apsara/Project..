

### ðŸ” Interpretation of Each Graph:

#### ðŸ“‰ 1. train/box\_loss

**Meaning** : Measures how well the predicted bounding boxes fit the ground truth (during training).
**Observation**: The loss decreases steadily, which is good â€” your model is learning to localize better.

#### ðŸ“‰ 2. train/obj\_loss

* **Meaning**: Objectness loss during training â€” how well the model predicts whether an object exists in a box.
* **Observation**: Decreasing trend indicates your model is getting better at distinguishing objects from background.

#### ðŸ“‰ 3. train/cls\_loss

* **Meaning**: Classification loss for multi-class problems (e.g., distinguishing between 'person' vs 'car').
* **Observation**: Always **0** â€” thatâ€™s expected for **single-class detection** (e.g., just detecting 'human').

#### ðŸ“ˆ 4. metrics/precision

* **Meaning**: Of all the detections made, how many were correct (TP / (TP + FP)).
* **Observation**: Rises above **0.9**, which is **excellent** â€” the model is making mostly correct predictions.

#### ðŸ“ˆ 5. metrics/recall

* **Meaning**: Of all actual objects, how many were detected (TP / (TP + FN)).
* **Observation**: Also stabilizes above **0.85**, meaning it's catching most humans in images.

#### ðŸ“‰ **6. val/box\_loss**

* **Meaning**: Bounding box loss on the validation set.
* **Observation**: Follows the same downward trend as training â€” shows **good generalization**.

#### ðŸ“‰ **7. val/obj\_loss**

* **Meaning**: Objectness loss on the validation set.
* **Observation**: Decreases and stabilizes â€” again, a healthy sign.

#### ðŸ“‰ **8. val/cls\_loss**

* **Meaning**: Always 0 â€” again, **expected for single-class models**.

#### ðŸ“ˆ **9. metrics/mAP\_0.5**

* **Meaning**: Mean Average Precision (mAP) at Intersection over Union (IoU) = 0.5. Think of this as overall detection performance.
* **Observation**: Climbs steadily and hits **\~0.88**, which is **very good**.

#### ðŸ“ˆ **10. metrics/mAP\_0.5:0.95**

* **Meaning**: Stricter mAP, averaging performance across IoU thresholds from 0.5 to 0.95.
* **Observation**: Ends around **0.72**, which is **strong** â€” your model is not just detecting, it's doing it accurately and consistently.

### âœ… Summary

```python
import gradio as gr
import os
import pandas as pd
import shutil
import subprocess

# Paths (modify as needed)
YOLO_PATH = "/Users/pawandinendra/Desktop/model/human/yolov5"
WEIGHTS = os.path.join(YOLO_PATH, "runs/train/human_yolov53/weights/best.pt")
DETECT_FOLDER = os.path.join(YOLO_PATH, "runs/detect/gui_detect")

def detect_human(image):
    # Save uploaded image
    image_path = os.path.join(YOLO_PATH, "temp_input.jpg")
    image.save(image_path)

    # Run YOLOv5 detect.py
    command = [
        "python3", os.path.join(YOLO_PATH, "detect.py"),
        "--weights", WEIGHTS,
        "--img", "640",
        "--conf-thres", "0.01",
        "--source", image_path,
        "--save-conf",
        "--save-txt",
        "--project", os.path.join(YOLO_PATH, "runs/detect"),
        "--name", "gui_detect",
        "--exist-ok",
        "--augment"
    ]
    subprocess.run(command, stdout=subprocess.DEVNULL)

    # Output paths
    result_img = os.path.join(DETECT_FOLDER, "temp_input.jpg")
    label_file = os.path.join(DETECT_FOLDER, "labels/temp_input.txt")

    # Create results table
    table = "No humans detected."
    if os.path.exists(label_file):
        with open(label_file, "r") as f:
            lines = f.readlines()

        if lines:
            data = []
            for line in lines:
                parts = line.strip().split()
                class_id = int(parts[0])
                x_center, y_center, width, height, conf = map(float, parts[1:])
                data.append({
                    "Class": class_id,
                    "x_center": x_center,
                    "y_center": y_center,
                    "width": width,
                    "height": height,
                    "Confidence": round(conf, 2)
                })
            df = pd.DataFrame(data)
            table = df

    return result_img, table


# Gradio UI
demo = gr.Interface(
    fn=detect_human,
    inputs=gr.Image(type="pil", label="Upload Image"),
    outputs=[
        gr.Image(type="filepath", label="Detected Image"),
        gr.Dataframe(label="Detections")
    ],
    title="Human Detection with YOLOv5",
    description="Upload an image. The model will detect humans and show bounding boxes + confidence."
)

demo.launch()
```

2. Tkinter Desktop App (Optional GUI)

```python
import tkinter as tk
from tkinter import filedialog, Label, Button
from PIL import Image, ImageTk
import subprocess
import os

YOLO_PATH = "/Users/pawandinendra/Desktop/model/human/yolov5"
WEIGHTS = os.path.join(YOLO_PATH, "runs/train/human_yolov53/weights/best.pt")
DETECT_NAME = "tk_detect"
DETECT_FOLDER = os.path.join(YOLO_PATH, f"runs/detect/{DETECT_NAME}")

def detect(file_path):
    command = [
        "python3", os.path.join(YOLO_PATH, "detect.py"),
        "--weights", WEIGHTS,
        "--img", "640",
        "--conf-thres", "0.01",
        "--source", file_path,
        "--save-conf",
        "--save-txt",
        "--project", os.path.join(YOLO_PATH, "runs/detect"),
        "--name", DETECT_NAME,
        "--exist-ok"
    ]
    subprocess.run(command)

    out_path = os.path.join(DETECT_FOLDER, os.path.basename(file_path))
    return out_path

def browse_file():
    file_path = filedialog.askopenfilename()
    if file_path:
        result = detect(file_path)
        image = Image.open(result)
        image = image.resize((400, 300))
        img_tk = ImageTk.PhotoImage(image)
        panel.config(image=img_tk)
        panel.image = img_tk
        label_status.config(text=f"Detected: {os.path.basename(file_path)}")

# GUI
root = tk.Tk()
root.title("Human Detector")

panel = Label(root)
panel.pack(pady=10)

btn = Button(root, text="Select Image", command=browse_file)
btn.pack(pady=10)

label_status = Label(root, text="Select an image to detect")
label_status.pack()

root.mainloop()
```

```python
from flask import Flask, render_template, request, redirect, url_for
import os
import uuid
import subprocess
import shutil

app = Flask(__name__)

# Paths
YOLO_PATH = "/Users/pawandinendra/Desktop/model/human/yolov5"
WEIGHTS = os.path.join(YOLO_PATH, "runs/train/human_yolov53/weights/best.pt")
UPLOAD_FOLDER = "static/uploads"
RESULT_FOLDER = "static/results"
DETECT_NAME = "flask_detect"

os.makedirs(UPLOAD_FOLDER, exist_ok=True)
os.makedirs(RESULT_FOLDER, exist_ok=True)

@app.route("/", methods=["GET", "POST"])
def index():
    if request.method == "POST":
        file = request.files["image"]
        if file.filename != "":
            filename = f"{uuid.uuid4().hex}.jpg"
            upload_path = os.path.join(UPLOAD_FOLDER, filename)
            file.save(upload_path)

            # Run detection
            subprocess.run([
                "python3", os.path.join(YOLO_PATH, "detect.py"),
                "--weights", WEIGHTS,
                "--img", "640",
                "--conf-thres", "0.01",
                "--source", upload_path,
                "--save-conf",
                "--save-txt",
                "--project", os.path.join(YOLO_PATH, "runs/detect"),
                "--name", DETECT_NAME,
                "--exist-ok"
            ], stdout=subprocess.DEVNULL)

            # Move result to static/results/
            result_path = os.path.join(YOLO_PATH, "runs/detect", DETECT_NAME, filename)
            output_path = os.path.join(RESULT_FOLDER, filename)
            shutil.copy(result_path, output_path)

            return redirect(url_for("result", filename=filename))
    return render_template("index.html")

@app.route("/result/<filename>")
def result(filename):
    return render_template("result.html", filename=filename)
```
